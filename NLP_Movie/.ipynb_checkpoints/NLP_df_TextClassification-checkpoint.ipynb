## Carga de archivos en DF

```python
allDocs= [(index, nlp(row[text]), row[category]) for index, row in df2.iterrows()]
```

```python
for index, row in df.iterrows():
    nlp(row[text])
```

```python
allDocs2 = [(doc, ct, w.text.lower(),w.pos_,w.tag_,w.is_stop,w.lemma_,ps.stem(w.text)) 
            for doc, txt, ct in allDocs for w in txt]
```

```python
#DF por palabra
df = pd.DataFrame(allDocs2, columns = ['document','category','word','pos','tag','is_stop','lemma','stem'])
df.sample(10)
```

# ////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

## Top words

```python
Top_words = 1000
key = 'word'
#Exclude:
wordEX = []
posEx = ['PUNCT','SPACE']
tagEx = ['PRP']
is_stopEx = [True]
lemmaEx = []
stemEx = []
```

```python
df2 = df[~df.word.isin(wordEX) & ~df.is_stop.isin(is_stopEx) & ~df.pos.isin(posEx) & ~df.tag.isin(tagEx) & 
         ~df.lemma.isin(lemmaEx)& ~df.stem.isin(stemEx)]
```

```python
Top_w = df2.groupby([key])[key].count().sort_values(ascending=False)[0:Top_words]
Top_w.head(10)
```

```python
Top_w2 = Top_w.index.values
```

# ////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

# Features

```python
key2 = key+'s'
df3 = df.groupby(['document','category'])[key].apply(list).reset_index(name=key2 )
df3.head()
```

```python
def find_features(lst):                  # Para identificar las palabras seleccionadas
    features = {}
    for w in Top_w2: 
        features[w] = (w in lst)
        
    return features
```

```python
df3['features'] = df3[key2].apply(find_features)
df3['length'] = df3[key2 ].apply(len)
df3.head()
```

# ////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

# Classificaci√≥n

```python
#'NLTK Naive Bayes'
#'MultinomialNB'
#'BernoulliNB'
#'LogisticRegression'
#'SGDClassifier'
#'NuSVC'
#'SVC'
Method = 'SVC'
Training_pcg = 0.95
n = 30 # Iteraciones 
```

```python
m =round(df3.shape[0]*Training_pcg)
```

```python
if Method == 'MultinomialNB':
    from sklearn.naive_bayes import MultinomialNB
elif Method == 'BernoulliNB':
    from sklearn.naive_bayes import BernoulliNB
elif Method == 'LogisticRegression':
    from sklearn.linear_model import LogisticRegression
elif Method == 'SGDClassifier':
    from sklearn.linear_model import SGDClassifier
elif Method == 'NuSVC':
    from sklearn.svm import NuSVC
elif Method == 'SVC':
    from sklearn.svm import SVC
```

```python
acc = []
if Method != 'NLTK Naive Bayes':
    from nltk.classify.scikitlearn import SklearnClassifier
    exec('classifier = '+'SklearnClassifier('+Method+'())') 
    
for i in range(n):
    df3 = df3.sample(frac=1).reset_index(drop = True)
    training_set = df3.loc[:(m-1),['features','category']].as_matrix()
    testing_set = df3.loc[m:,['features','category']].as_matrix()
    
    if Method == 'NLTK Naive Bayes':
        classifier = nltk.NaiveBayesClassifier.train(training_set)

    if Method != 'NLTK Naive Bayes':
        classifier.train(training_set)

    acc.append(nltk.classify.accuracy(classifier,testing_set))
    
avg = sum(acc)/len(acc)
```

```python
print(Method +" accuracy percent:",f'{round(avg*100,2)}%')
```
